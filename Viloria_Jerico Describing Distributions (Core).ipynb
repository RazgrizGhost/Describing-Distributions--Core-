```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("medical_dataset.csv")
display(df.head())
df.info()
```

# VitD_levels

```python
vit_mean = df["VitD_levels"].mean()
vit_median = df["VitD_levels"].median()
vit_plus = vit_mean + df["VitD_levels"].std()
vit_minus = vit_mean - df["VitD_levels"].std()

fig,ax = plt.subplots(figsize=(12,6))
sns.histplot(df['VitD_levels'], ax = ax, stat = 'density', kde = True)
ax.axvline(vit_mean, color = 'r', label = f"Mean = {vit_mean}")
ax.axvline(vit_median, color = 'g', ls = "--", label = f"Median = {vit_median}")
ax.axvline(vit_plus, color = "k", label = f"+1 std = {vit_plus}")
ax.axvline(vit_minus, color = "k", label = f"-1 std = {vit_minus}")
ax.axvspan(vit_plus, vit_minus, color = "y", zorder = 0)
ax.legend();
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12,6))
sns.histplot(df["VitD_levels"], ax = axes[0])
sns.scatterplot(df["VitD_levels"], ax = axes[1])
```

## Answer the following questions

- Is it Discrete or Continuous?
    - VitD_levels is continuous.
- Does it have a skew? If so, which direction (+/-)
    - It skews in the negative direction.
- What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)
    - It displays a Mesokurtic kurtosis.
